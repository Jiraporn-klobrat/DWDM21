## flights

### ตรวจสอบค่า Missing ของตาราง flights

````python
flights.isnull() # ตรงไหนเป็นช่องว่างจะขึ้นว่า True ตรงไหนไม่เป็นช่องว่างจะขึ้นว่า False
````

````python
flights.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง สรุปคือทุกคอลัมน์ไม่มีค่า  missing
````

### เชื่อมตาราง

````python
merged_table = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน')
merged_table
````

### ตรวจสอบค่า Missing ของตาราง merged_table

````python
merged_table.isnull() # ตรงไหนเป็นช่องว่างจะขึ้นว่า True ตรงไหนไม่เป็นช่องว่างจะขึ้นว่า False
````

````python
merged_table.isnull().any() # ดูว่าคอลัมน์ไหนมีค่า missing บ้าง สรุปคือทุกคอลัมน์ไม่มีค่า  missing
````

## จำนวนเครื่องบินขาเข้ามากที่สุด

````python
merged_table[['สายการบิน' ,	'รหัสสนามบินต้นทาง' , 'รหัสสนามบินปลายทาง' , 'รัฐ']] #ชี้เฉพาะคอลัมน์ที่ต้องการ
````

# 5. วันธรรมดา หรือวันเสาร์-อาทิตย์มีเครื่องบินขาเข้ามากกว่ากัน (วันจันทร์ = 1, วันอังคาร = 2, ...)

````python
merged_table = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน')
merged_table
````

````python
merged_table[['วันในสัปดาห์' ,	'สายการบิน' , 'ความล่าช้าขาออก' , 'ความล่าช้าขาเข้า']] #ชี้เฉพาะคอลัมน์ที่ต้องการ
````

````python
merged_table = merged_table.sort_values(by=['วันในสัปดาห์'])   # เรียงข้อมูลตามวันในสัปดาห์
merged_table
````

````python
merged_table[['รหัสสนามบินปลายทาง','วันในสัปดาห์']]
````

````python
max = merged_table.groupby('วันในสัปดาห์').max().sort_values(by=['รหัสสนามบินปลายทาง'],ascending=False)
max.iloc[:,3:4]
````
